# Outcomes and ∆Q 

Outcomes arise when you have a system where there is a _starting observable_ that, once observed, should lead to a _terminating observable_. 

Outcomes can fail (they are started but do not have a clear end, they _never_ occur) or they have a duration over which they are occurring.

Outcomes can be composed sequentially or run concurrently - with either the first or last to finish as they composite completion criterion.

```haskell
o1, o2, o3 :: DQ
o1 = wait 0.5
o2 = wait 1.0
o3 = wait 2.0
```

```haskell
o4 = o1 .>>. o2
o5 = o1 ./\. o3
o6 = o1 .\/. o3
```

```haskell
("o1", asFloat <$> deadline o1)
("o2", asFloat <$> deadline o2)
("o3", asFloat <$> deadline o3)
("o4", asFloat <$> deadline o4)
("o5", asFloat <$> deadline o5)
("o6", asFloat <$> deadline o6)
```

What we have created here is a deterministic world. 
Outcomes either _occur_ or are _abandoned_, when they do occur they occur at a deterministic time. 
Although, through `firstToFinish` (`.\/.`) and `lastToFinish` (`./\.`) we can express forms of concurrency, the results are fixed scalar values.
Even if the resulting outcome never terminates.

```haskell
asFloat <$> deadline ( o1 .>>. never)
```

It is rare for the real world to be that deterministic. 
It is more usual for outcomes to have uncertainity, to some extent.

```haskell
o7 :: DQ
o7 = uniform 1 2
```

Now things start to get more interesting, instead of the determinism we now enter the realm of possibilities and probabilities. 
Note that this is not (neceessarily) a world of uncertainty, but one of predictablity with bounds (and, as we will see, distributions) on durations and probablity.

```haskell
("earliest o7", asFloat <$> earliest o7)
("deadline o7", asFloat <$> deadline o7)
```

in such a realm the questions that can be asked are more interesting.
What if the outcome of interest was `o7` followed by another `o7`?
What would the bounds be?
What would the median result be?
What would the 90th centile be?
What would the probability of occurance with 3.5 seconds?

```haskell
o8 = o7 .>>. o7
("bounds o8", asFloat <$> earliest o8, asFloat <$> deadline o8)
("median o8", asFloat <$> quantile 0.5 o8)
("90th centile o8", asFloat <$> quantile 0.9 o8)
("prob. within 3.5s", asFloat $ successWithin o8 3.5)
```

Numbers are useful, but often you want to visualise such resulting properties.

```haskell
toRenderable $ plotCDF "Visualising o8" o8
```

An you may want to add centiles to the visualisation.

```haskell
toRenderable $ plotCDFWithCentiles "o8 with centiles"  [0.25, 0.5, 0.75, 0.9] o8
```

Up to this point we've been dealing with things that either definitly (if eventaually) occur or never occur. 
It is as this point we step through the looking glass and incorporate the notion may be a mixture of things that take time to occur along with a possibility of not occuring. 
We are going to bring together both notion of taking time and 'failing' (or some other notion of non-occurance)

```haskell
o9 :: DQ
o9 = choice 0.75 o8 never

("bounds o9", asFloat <$> earliest o9, asFloat <$> deadline o9)
("median o9", asFloat <$> quantile 0.5 o9)
("90th centile o9", asFloat <$> quantile 0.9 o9)
("prob. within 3.5s", asFloat $ successWithin o9 3.5)
```

The bounds on the values (which can occur) have not changed, but the meadian value has. There is no 90th centile any more and the probability of completing with 3.5 seconds is reduced.


```haskell
toRenderable $ plotCDFWithCentiles "o9 with centiles"  [0.25, 0.5, 0.75, 0.9] o9
```

Can we create more certainty of (at least some form of) and outcome?
What if we introduce a first-to-finish with some timeout? say 5s?

```haskell
o10 = o9 `firstToFinish` wait 5
("bounds o10", asFloat <$> earliest o10, asFloat <$> deadline o10)
("median o10", asFloat <$> quantile 0.5 o10)
("90th centile o10", asFloat <$> quantile 0.9 o10)
("prob. within 3.5s", asFloat $ successWithin o10 3.5)
toRenderable $ plotCDFWithCentiles "o10 with centiles"  [0.25, 0.5, 0.75, 0.9] o10
```

We have recovered the 'certainity', but we had to do that by inspection and insertion of specific values.
We don't have constrain ourselves that way, we are working in a higher order framework here - we can inspect the system we are reasoning about and create suitable mitigation strategies.

Imagine we wanted to ensure that the system always made progress, even in the presence of failure. We might ensure that this happens by encoding the 'rule' that, if an outcome can fail then we will create a 'timeout' that occurs after the latest time.

More than this, we will create such a timeout so that it 'fires' at a specified fraction of the active duration of the outcome (the active duration being the `deadline - earliest`).

```haskell
ensureProgress :: DeltaQ o => Duration o -> o -> o
ensureProgress f o 
  = o `firstToFinish` timeguard (earliest o, deadline o)
  where
    timeguard (Occurs a, Occurs b)
      = wait $ b + (b - a) * f
    timeguard _ = never
```

```haskell
o11 = ensureProgress 0.25 o9
toRenderable $ plotCDFWithCentiles "o11 with centiles"  [0.25, 0.5, 0.75, 0.9] o11
```
